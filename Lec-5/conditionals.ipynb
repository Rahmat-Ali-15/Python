## Conditionals
##### if-else

```python
# Check if a person can vote or not

# take input from the user
age = int(input("Enter the age : "))

# decision making
if age>= 18:
    print(f"You can vote Since you are {age} years old")
else:
    print(f"You cannot vote Since you are {age} years old")
```

```python
# Check whether a number is even or odd

number = int(input("Enter the number :"))

if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
```

```python
# Check if a number is positive, negative or zero

n = int(input("Enter the number : "))

if n<0:
    print(f"{n} is Negative")
elif n > 0:
    print(f"{n} is Positive")
else:
    print(f"{n} is Zero")
```

```python
n = int(input("Enter the number : "))

if n > 0:
    if n % 2 == 0:
        print(f"{n} -> Even Positive")
    else:
        print(f"{n} -> Odd Positive")
elif n < 0:
    if n % 2 == 0:
        print(f"{n} -> Even Negative")
    else:
        print(f"{n} -> Odd Negative")
else:
    print(f"{n} -> Zero")
```
