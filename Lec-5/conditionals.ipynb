## Conditionals
##### if-else clause

```python
# Check if a person can vote or not

# take input from the user
age = int(input("Enter the age : "))

# decision making
if age>= 18:
    print(f"You can vote Since you are {age} years old")
else:
    print(f"You cannot vote Since you are {age} years old")
```

```python
# Check whether a number is even or odd

number = int(input("Enter the number :"))

if number % 2 == 0:
    print(f"{number} is Even")
else:
    print(f"{number} is Odd")
```

```python
# Check if a number is positive, negative or zero

n = int(input("Enter the number : "))

if n<0:
    print(f"{n} is Negative")
elif n > 0:
    print(f"{n} is Positive")
else:
    print(f"{n} is Zero")
```

```python
n = int(input("Enter the number : "))

if n > 0:
    if n % 2 == 0:
        print(f"{n} -> Even Positive")
    else:
        print(f"{n} -> Odd Positive")
elif n < 0:
    if n % 2 == 0:
        print(f"{n} -> Even Negative")
    else:
        print(f"{n} -> Odd Negative")
else:
    print(f"{n} -> Zero")
```

### Grading Symtem with Marks

```python
# Grading system according to marks

n = int(input("Enter the marks : "))

if n >= 90 <= 100:
    print(f"Your grade is A")
elif n >= 80 > 90:
    print(f"Your grade is B")
elif n >= 70 > 80:
    print(f"Your grade is C")
elif n >= 60 > 70:
    print(f"Your grade is D")
else:
    print(f"Fail")
```

### Grading System with Percentage

```python
# Grading system according to percentage

total_marks = 500

obtain_marks = float(input("Enter the obtain percentage : "))

percentage = (obtain_marks * 100) / total_marks

if percentage >= 90 <= 100:
    print(f"Grade: A")
elif percentage >= 80 > 90:
    print(f"Grade: B")
elif percentage >= 70 > 80:
    print(f"Grade: C")
elif percentage >= 60 > 70:
    print(f"Grade: D")
else:
    print(f"Fail")
```
