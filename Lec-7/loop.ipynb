```python
# User Input 
start = int(input("Enter the start value : "))
end = int(input("Enter the end value : "))

count = start # Initialization  

while count < end + 1:
    print(f"Current Value : {count}\nStoping : {end}\nCondition : {count} <= {end} --> {count <= end}\n")

    # print(count, end = ' ')  
    count += 1 # Updation 
```

```python
# Print even number between 1 to 10

start = 1
stop = 10

while start <= stop:
    if start % 2 == 0:
        print(start, end = " ")
        start += 1
    else:
        start += 1
```

```python
# Print the sum of numbers form 1 to 5.

start = 1
stop = 5
sum = 0

while start <= stop:
    sum += start
    start += 1
print(f"The sum of number form 1 to {stop} is {sum}")
```

```python
# Print the sum of n natural numbers

start = 1
stop = int(input("Enter the number : "))
sum  = 0

while start <= stop:
    sum = sum + start
    start += 1
print(sum)
```

```python
# Print the table of 2 from 2 x 1 to 2 x 10.

start = 1
end = 10

table = 2

while start <= end:
    result = table * start
    print(f"{table} x {start} = {result}")
    start += 1
```
