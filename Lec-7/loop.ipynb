```python
# Print number form 1 to 10
start = 1
stop = 10

while start <= stop:
    print(f"Current Value : {start}\nStop Value : {stop}\nCondition : {start} <= {stop} --> {start<= stop}\n")
    start +=1
else:
    print(f"Current Value : {start}\nStop Value : {stop}\nCondition : {start} <= {stop} --> {start<= stop}\n")
```

```python
# Print number from 10 to 1

start = 10
end = 1

count = start

while count>= end:
    print(count)
    count -= 1
```

```python
# Print even number between 1 to 10

start = 1
stop = 10

while start <= stop:
    if start % 2 == 0:
        print(start, end = " ")
        start += 1
    else:
        start += 1
```
