```python
# Print number form 1 to 10
start = 1
stop = 10

while start <= stop:
    print(f"Current Value : {start}\nStop Value : {stop}\nCondition : {start} <= {stop} --> {start<= stop}\n")
    start +=1
else:
    print(f"Current Value : {start}\nStop Value : {stop}\nCondition : {start} <= {stop} --> {start<= stop}\n")
```

```python
# Print number from 10 to 1

start = 10
end = 1

count = start

while count>= end:
    print(count)
    count -= 1
```

```python
# Print even number between 1 to 10

start = 1
stop = 10

while start <= stop:
    if start % 2 == 0:
        print(start, end = " ")
        start += 1
    else:
        start += 1
```

```python
# Print the sum of numbers form 1 to 5.

start = 1
stop = 5
sum = 0

while start <= stop:
    sum += start
    start += 1
print(f"The sum of number form 1 to {stop} is {sum}")
```

```python
# Take a number from the user and find the sum of all numbers from 1 to that number.

start = 1
stop = int(input("Enter the last number : "))
sum = 0

while start<= stop:
    sum += start
    start += 1
print(f"The sum of Numbers from 1 to {stop} is {sum}")
```

```python
# Print the table of 2 from 2 x 1 to 2 x 10.

start = 1
end = 10

table = 2

while start <= end:
    result = table * start
    print(f"{table} x {start} = {result}")
    start += 1
```

```python
# Take a number from the user and print its multiplication table.

start = 1
stop = 10

table = int(input("Enter the number which table will be printed :"))

while start<= stop:
    result = start * table
    print(f"{table} x {start} = {result}")
    start += 1
```
