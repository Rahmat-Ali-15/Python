## Assignment Operator

```python

a = 10
b = 20

print(f" a = {a}, b = {b}")
a += b

print(f"a= {a}, b = {b}")
a -= b

print(f"a= {a}, b = {b}")
a *= b

print(f"a= {a}, b = {b}")
a /= b

print(f"a= {a}, b = {b}")
a //= b

print(f"a= {a}, b = {b}")
a %= b

print(f"a= {a}, b = {b}")
```

## Operators :  Bitwise

```python
a = 25
b = 20

print(a & b)
print(a | b)
print(a << 1)
print(b >> 2 )
print(~a)
```
