##### Arithematic Operators

```python

a = 12
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
```

##### Comparison Operators

```python

a = 10
b = 10

print(f"a = {a}, b = {b}")
print(f"{a} == {b} : {a == b}")
print(f"{a} != {b} : {a != b}")
print(f"{a} < {b} : {a < b}")
print(f"{a} > {b} : {a > b}")
print(f"{a} <= {b} : {a <= b}")
print(f"{a} >= {b} : {a >= b}")
```
